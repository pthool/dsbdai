```python
import pandas as pd
from scipy import stats
```

```python
heart=pd.read_csv('heart.csv')
heart.head()
```

```python
#DATA TRANSFORMATION
from sklearn.preprocessing import LabelEncoder
scaler=LabelEncoder()
heart['patient_id']=scaler.fit_transform(heart['patient_id'])

heart.loc[heart['chol']<200, 'chol']=heart['chol'].mean()
heart.head()
```

```python
#ERROR HANDLING (ZSCORES - DISTANCE FROM STANDARD DEVIATION)
zscore=stats.zscore(heart['trestbps'])
heart_no_outliers=heart[(zscore<3)]
heart
```

```python
#data transformation and DATA MODEL BUILDING

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor

x=heart.iloc[:, :-1]
y=heart['target']
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=42)

scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)

model=RandomForestRegressor()
model.fit(x_train_scaled, y_train)

accuracytrained=model.score(x_train_scaled, y_train)
accuracytest=model.score(x_test_scaled, y_test)

print("accuracy of trained model: ", accuracytrained)
print("accuracy of test model: ", accuracytest)
```
