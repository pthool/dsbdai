```python
import pandas as pd
```

```python
heart=pd.read_csv('heart.csv')
heart.head()
```

```python
air=pd.read_csv('AirQuality.csv', sep=';')
air.head()
```

```python
Date=air.loc[:, 'Date']
Date.head()
```

```python
heart=pd.merge(heart, Date, left_index=True, right_index=True)
heart.head()
```

```python
merged=pd.merge(heart, air, on='Date')
merged.head()
```

```python
merged=merged.drop(['Unnamed: 15', 'Unnamed: 16'], axis=1)
merged.head()
```

```python
merged=merged.dropna()
merged.shape
```

```python
merged=merged.drop_duplicates()
merged.shape
```
