```python
import pandas as pd
```

```python
heart=pd.read_csv('heart.csv')
air=pd.read_csv('AirQuality.csv', sep=';')
```

```python
heart.head()
```

```python
#Data Cleaning
heart.dropna()
```

```python
heart.drop_duplicates()
```

```python
air.dropna()
```

```python
air.drop_duplicates()
```

```python
heart.isnull().sum()
```

```python
heart=heart.drop(['patient_id'],axis=1)
heart.isnull().sum()
```

```python
air.isnull().sum()
```

```python
air=air.drop(['Unnamed: 15', 'Unnamed: 16'], axis=1)
air.isnull().sum()
```

```python
#DATA INTEGRATION
air.head()
date=air.loc[:,['Date']]
heart=pd.merge(date, heart, left_index=True, right_index=True)
heart.head()
```

```python
merged_df=pd.merge(heart, air, on='Date')
merged_df.head()
```

```python
merged_df.columns.tolist()
```

```python
merged_df=merged_df.drop(['T','CO(GT)',
 'PT08.S1(CO)',
 'NMHC(GT)',
 'C6H6(GT)'], axis=1)
merged_df.columns.tolist()
```

```python
merged_df.shape
merged_df.head()
```

```python
#DATA TRANSFORMATION
from sklearn.preprocessing import LabelEncoder
scaler=LabelEncoder()
merged_df['chol']=scaler.fit_transform(merged_df['chol'])

merged_df.loc[merged_df['NO2(GT)']<100, 'NO2(GT)']=merged_df['NO2(GT)'].mean()
merged_df.tail()
```
