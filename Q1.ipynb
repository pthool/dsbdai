```python
import pandas as pd
```

```python
df=pd.read_csv('dataset_Facebook.csv', sep=';')
```

```python
#A) Create Data Subsets

df_subset1 = df.loc[0:4, ['like', 'share', 'comment']]
df_subset1.head()
```

```python
df_subset2 = df.iloc[5:10, [2, 3, 4]]
df_subset2.head()
```

```python
df_subset3=df['like'].div(df['share']).to_frame('like-share-ratio')
df_subset3.head()
```

```python
#B) Merge Data

df_merged=pd.merge(df, df_subset3, left_index=True, right_index=True)
df_merged.head()
```

```python
df_transposed=df.transpose()
print(df_transposed)
```
